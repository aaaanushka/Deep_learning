```python
# import libraries
import numpy as np
import tensorflow as tf
from tensorflow.keras.preprocessing.text import Tokenizer
from tensorflow.keras.preprocessing.sequence import pad_sequences
from tensorflow.keras.models import Sequential
from tensorflow.keras.layers import Embedding, LSTM, Dense

print("TensorFlow version:", tf.__version__)
```

```python
# enter corpus and some detail about them
corpus = [
    "deep learning models can solve complex problems",
    "machine learning is transforming the world",
    "artificial intelligence is the future of technology",
    "data science helps in making better decisions",
    "neural networks are inspired by the human brain",
    "python is widely used for machine learning",
    "big data requires efficient processing techniques",
    "supervised learning uses labeled data",
    "unsupervised learning finds hidden patterns",
    "reinforcement learning learns through rewards",

    "natural language processing enables computers to understand text",
    "computer vision allows machines to interpret images",
    "algorithms improve with more training data",
    "training a model requires clean and relevant data",
    "overfitting happens when a model memorizes training data",
    "regularization techniques help reduce overfitting",
    "gradient descent is used to minimize loss",
    "loss functions measure prediction errors",
    "deep learning models need large datasets",
    "feature engineering improves model performance",

    "data preprocessing is an important step in machine learning",
    "models should be evaluated using test data",
    "accuracy is not always the best evaluation metric",
    "precision and recall are useful for classification problems",
    "f1 score balances precision and recall",
    "cross validation improves model reliability",
    "hyperparameter tuning improves model performance",
    "neural networks consist of multiple layers",
    "activation functions introduce non linearity",
    "relu is a commonly used activation function",

    "sigmoid function outputs values between zero and one",
    "softmax function is used for multi class classification",
    "embedding layers convert words into dense vectors",
    "lstm networks are useful for sequence prediction",
    "rnn models can handle sequential data",
    "transformers are more powerful than rnn models",
    "attention mechanism improves model performance",
    "bert is a popular transformer based model",
    "gpt models are used for text generation",
    "language models learn patterns from text data",

    "ai is used in healthcare for diagnosis",
    "ai is used in finance for fraud detection",
    "ai is used in education for personalized learning",
    "self driving cars use computer vision",
    "chatbots use natural language processing",
    "recommendation systems suggest relevant content",
    "voice assistants understand spoken language",
    "data visualization helps in understanding trends",
    "cloud computing supports scalable machine learning",
    "edge computing enables real time predictions"
]

print("Number of sentences:", len(corpus))
print("\nSample corpus:")
for line in corpus:
    print("-", line)
```

```python
# step 2 tokenizers: convert words вҶ’ numbers

tokenizer = Tokenizer()
tokenizer.fit_on_texts(corpus)
word_index = tokenizer.word_index
total_words = len(tokenizer.word_index) + 1
print(total_words)
print("\nWord index:")
print(word_index)
```

```python
# convert sentences into sequences
input_sequences = []

for line in corpus:
    token_list = tokenizer.texts_to_sequences([line])[0]

    for i in range(1, len(token_list)):
        n_gram = token_list[:i+1]
        input_sequences.append(n_gram)
print("Total training sequences:", len(input_sequences))
print("\nSome sequences before padding:")
for seq in input_sequences[:10]:
    print(seq)
```

```python
# step 4: padding-> make equal length because neural networks need fixed-size input

max_seq_len = max(len(x) for x in input_sequences)

input_sequences = pad_sequences(input_sequences, maxlen=max_seq_len, padding='pre')
print("Maximum sequence length:", max_seq_len)
print("\nPadded sequences:")
print(input_sequences[:10])
```

```python
# STEP 5: Split into X and y
X = input_sequences[:, :-1]
y = input_sequences[:, -1]
# STEP 6: One-hot encoding of output because Output is a word вҶ’ we convert to probability distribution
y = tf.keras.utils.to_categorical(y, num_classes=total_words)

print("Shape of X:", X.shape)
print("Shape of y:", y.shape)
```

```python
# STEP 7: Build LSTM Model
from tensorflow.keras.layers import Dropout
model = Sequential()

model.add(Embedding(total_words, 50, input_length=max_seq_len-1))
model.add(LSTM(80))   # вҶ“ reduced from 150/100
model.add(Dropout(0.3))
model.add(Dense(total_words, activation='softmax'))

model.compile(loss='categorical_crossentropy', optimizer='adam', metrics=['accuracy'])

model.summary()
```

```python
# STEP 8: Train model
history = model.fit(X, y, epochs=100, verbose=1)


```

```python
# STEP 10: Prediction function
def predict_next_word(seed_text, next_words):
    for _ in range(next_words):
        token_list = tokenizer.texts_to_sequences([seed_text])[0]
        token_list = pad_sequences([token_list], maxlen=max_seq_len-1, padding='pre')

        predicted = np.argmax(model.predict(token_list), axis=-1)

        for word, index in tokenizer.word_index.items():
            if index == predicted:
                seed_text += " " + word
                break

    return seed_text
```

```python
print(predict_next_word("machine learning is", 3))
print(predict_next_word("deep learning", 3))
print(predict_next_word("data science", 3))
```
