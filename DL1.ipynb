OPERATION IN TENSOR (EXPERIMENT- 3)

```python
# operation in scalar
v= tf.constant(10)
u= tf.constant(10)
print(v+u)
print(v-u)
print(v*u)
print(v/u)
print(v^u)

```

```python
# operation in vector
a=tf.constant([10,20,30])
b = tf.constant([50,60,70])
print(a+b)
print(a-b)
print(a*b)
print(a/b)
print(a^b)
```

```python
#operation on 2d
m = tf.constant([[1,2],[3,4]])
n = tf.constant([[5,6],[7,8]])
print(m+n)
print(m-n)
print(m*n)
print(m/n)
print(m^n)
```

Datatype and shape of tensor(EXPERIMENT_4)

```python
import tensorflow as tf

# Scalar (0D Tensor)
scalar = tf.constant(5)

# Vector (1D Tensor)
vector = tf.constant([10, 20, 30])

# Matrix (2D Tensor)
matrix = tf.constant([[1, 2, 3],
                      [4, 5, 6]])

# Scalar details
print("Scalar Tensor:", scalar)
print("Data type:", scalar.dtype)
print("Shape:", scalar.shape)
print("Number of dimensions:", scalar.ndim)
print()

# Vector details
print("Vector Tensor:", vector)
print("Data type:", vector.dtype)
print("Shape:", vector.shape)
print("Number of dimensions:", vector.ndim)
print()

# Matrix details
print("Matrix Tensor:", matrix)
print("Data type:", matrix.dtype)
print("Shape:", matrix.shape)
print("Number of dimensions:", matrix.ndim)
```

NUumpy array to Tenser(EXPERIMENT-5)

```python
import numpy as np

numpy_scalar = np.array(5)
tensor_scalar = tf.convert_to_tensor(numpy_scalar)

print("----- SCALAR -----")
print("\nNumPy Scalar:", numpy_scalar)
print("\nTensor Scalar:", tensor_scalar)



numpy_vector = np.array([1, 2, 3, 4])
tensor_vector = tf.convert_to_tensor(numpy_vector)


print("\nNumPy Vector:", numpy_vector)
print("\nTensor Vector:", tensor_vector)



numpy_matrix = np.array([[1, 2, 3],
                         [4, 5, 6]])
tensor_matrix = tf.convert_to_tensor(numpy_matrix)


print("\nNumPy Matrix:")
print(numpy_matrix)
print("\nTensor Matrix:")
print(tensor_matrix)
```
